# Linking it situ phenological observations with climate data

The following example leverages the [NEON Data](http://data.neonscience.org/data-api) and [Google Earth Engine](https://developers.google.com/earth-engine/) APIs to link in situ phenological observations with climate data.

We'll use the `requests` module to handle sending and receiving data from the NEON Data API and the `ee` module to handle calls to the Earth Engine API. Along the way, we'll invoke R for visualization purposes (i.e., using [`ggplot2`](https://github.com/tidyverse/ggplot2) and [`leaflet`](https://github.com/rstudio/leaflet) packages). We'll also define some variables to use a few times.

```python
%load_ext rpy2.ipython
```

```python
import requests
import json
import ee
from pandas.io.json import json_normalize
import pandas as pd


ee.Initialize()  # initialize the Earth Engine object.


PRODUCTCODE = "DP1.10055.001"  # the product code for plant phenology observations
SERVER = "http://data.neonscience.org/api/v0/"  # the current server address
```

First, we'll send a request to NEON's `/products` endpoint to get some information about what's available there. The response is in JSON format, which we can view using `json.dumps` for formatting.

```python
product_response = requests.get(SERVER + 'products/' + PRODUCTCODE)
product_response_json = product_response.json()
# print(json.dumps(product_response_json, indent=2))
```

We'll grab the information available under `data -> siteCodes` to determine which sites have generated phenology data. The `locations/` endpoints contains the relevant location information.

```python
data_products = product_response_json['data']['siteCodes']
site_codes = [d['siteCode'] for d in data_products]
# print(site_codes)
```

Let's pull location information (i.e., longitude/latitude) for each site. 

```python
metadata_fields = ['siteCode', 'locationDescription', 'locationElevation', 
                   'locationDecimalLongitude', 'locationDecimalLatitude']

site_metadata = []
for site_code in site_codes:
    location_response = requests.get(SERVER + 'locations/' + site_code)
    location_response_json = location_response.json()
#     print(json_normalize(location_response_json['data']).columns.values)
    site_metadata.append(json_normalize(location_response_json['data']).loc[:, metadata_fields])
site_metadata = pd.concat(site_metadata, axis=0)
# print(site_metadata)
```

```python
site_summary_pd_df = pd.DataFrame(pheno_obs_counts)

ee_features = []
for i in range(site_summary_pd_df.shape[0]):
    point_data = site_summary_pd_df.iloc[i, :]
    point = ee.Geometry.Point(list(point_data.loc[['lon', 'lat']]))
    ee_feature = ee.Feature(point, {'siteCode': point_data.loc['siteCode']})
    ee_features.append(ee_feature)
ee_feature_coll = ee.FeatureCollection(ee_features)
# print(ee_feature_coll.getInfo())


SIx_asset_path = 'users/LZachmann/SI-x_v6_0_0'
SIx = ee.ImageCollection(SIx_asset_path)
# print(SIx.size().getInfo()) 

def extract_six_to_point(image):
    image = image.addBands(ee.Image.constant(image.get('year')))
    timestep_fc = image.reduceRegions(
        collection=ee_feature_coll, reducer=ee.Reducer.first(), scale=250)
    return timestep_fc 
ee_results = SIx.map(extract_six_to_point).flatten().getInfo()
# print(ee_results)

# Convert the results of the reduction in EE back into a pandas DataFrame.
ee_results_pd_df = json_normalize(ee_results['features'])
col_suffixes = ['constant', 'b1', 'b2', 'siteCode']
selected_cols = ['properties.' + x for x in col_suffixes]
trunc_results = ee_results_pd_df.ix[:, selected_cols]
```

Visualize the results back in R.

```python
%%R -i trunc_results,col_suffixes

library(ggplot2)


names(trunc_results) <- col_suffixes
trunc_df <- trunc_results %>%
    reshape2::melt(id.vars=c('constant', 'siteCode')) %>%
    mutate(variable=ifelse(variable=='b1', 'First leaf', 'First bloom'),
           variable=factor(variable, levels=c('First leaf', 'First bloom'))) %>%
    tbl_df
    
ggplot(trunc_df, 
       aes(x=constant, y=value, group=siteCode, color=siteCode)) +
    geom_line() +
    labs(x='Year', y='Day of year') +
    facet_grid(.~variable)
```
